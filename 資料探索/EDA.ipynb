```python
import pandas as pd
```

```python
dflist = []
for i in range(1,18):
    for j in range(i+1,18):
        l1 = "0" + str(i) if i<10 else str(i)
        l2 = "0" + str(j) if j<10 else str(j)
        df1 = pd.read_csv(f"../模型訓練程式碼/train_combine_groupby/{l1}_traingroupby.csv")
        df2 = pd.read_csv(f"../模型訓練程式碼/train_combine_groupby/{l2}_traingroupby.csv")
        df1.drop(columns = ["Serial"],inplace = True)
        df = df1.merge(
            df2,
            on = "TimeGroup",
            suffixes=[f"-1",f"-2"],
            how = "inner"
        )
        dflist.append(df)
        
```

```python
df = pd.concat(dflist,axis=0)
df.to_csv("twopoint_scatter.csv",index=False)
```
