```python
import pandas as pd 
import json
```

##### 本檔案為整理不同的訓練方式在做交叉驗證的MAE比較結果彙整

```python
model_folder_name = "train_use_full_data_update_weather_data"  
with open(f"./model_each_location/{model_folder_name}/d.json", "r", encoding="utf-8") as file:
    mae_data = json.load(file)
```

```python
df1 = pd.json_normalize(mae_data,sep="-").transpose().reset_index(drop=False)
df1.columns = ["模型","MAE"]
df1["類別"] = "異地同時"
df1["模型"] = df1["模型"].str.split("-")
df1["模型"] = df1["模型"].str[1]
df1["裝置"] = df1["模型"].str.slice(start=3, stop=5)
df1
```

```python
model_folder_name = "train_use_full_data_update_weather_datasingle_location"  
with open(f"./model_each_location/{model_folder_name}/d.json", "r", encoding="utf-8") as file:
    mae_data = json.load(file)
```

```python
df2 = pd.json_normalize(mae_data).transpose().reset_index(drop=False)
df2.columns = ["模型","MAE"]
df2["類別"] = "同地同時"
df2["裝置"] = df2["模型"].str.split("_")
df2["裝置"] = df2["裝置"].str[0]
df2
```

```python
model_folder_name = "train_lastday_gridsearch_usealldata"  
with open(f"./model_each_location/{model_folder_name}/d_lastday.json", "r", encoding="utf-8") as file:
    mae_data = json.load(file)
```

```python
df3 = pd.json_normalize(mae_data).transpose().reset_index(drop=False)
df3.columns = ["模型","MAE"]
df3["類別"] = "同地異時"
df3["裝置"] = df3["模型"].str.split(".")
df3["裝置"] = df3["裝置"].str[0]
df3
```

```python
df = pd.concat([df1,df2,df3],axis=0)
df.to_csv("modelmae.csv",index=False)
```
