```python
import pandas as pd
```

#### 本檔案為將各個裝置1/1~10/31的資料與中央氣象局的資料以及自己計算之太陽仰角等資料做串接之程式碼
##### 結果檔可至Google Drive下載weather_sun_combine.csv

```python
sun_df = pd.read_csv("../太陽仰角與日射量資料計算整理/sunlight_data.csv")
sun_df["location"] = sun_df["location"].astype(str).str.zfill(2)
sun_df["DateTime"] = pd.to_datetime(sun_df["DateTime"])
["太陽仰角", "日射量 (W/m²)","太陽方位角", "太陽能板日射量 (W/m²)"]
sun_df
```

```python
weather_df_10mins = pd.read_csv("../資料前處理以及第三方資料/10mins_weather_data.csv")
weather_df_10mins["DateTime"] =  pd.to_datetime(weather_df_10mins["DateTime"])
# weather_df_10mins["日期"] = pd.to_datetime(weather_df_10mins["DateTime"].dt.date)
# weather_df_10mins["時間"] = weather_df_10mins["DateTime"].dt.hour.astype(str).str.zfill(2)
# weather_df_10mins["時段"] = weather_df_10mins["DateTime"].dt.minute.astype(str).str.zfill(2)

weather_df_10mins.fillna(0,inplace=True)
```

```python
weather_df_10mins
```

```python
weather_df_1hour = pd.read_csv("../資料前處理以及第三方資料/station466990weather.csv")
weather_df_1hour = weather_df_1hour.loc[:,["DateTime","TX01","GR01","RH01","UV01"]]
weather_df_1hour["DateTime"] = weather_df_1hour["DateTime"].astype(str)
weather_df_1hour["日期"] = pd.to_datetime(weather_df_1hour["DateTime"].str.slice(0,8))
weather_df_1hour["時間"] = weather_df_1hour["DateTime"].str.slice(8,10)
weather_df_1hour.loc[weather_df_1hour["時間"]=="24","時間"] = "00"
weather_df_1hour.fillna(0,inplace=True)
weather_df_1hour.drop_duplicates(subset=["日期","時間"],inplace=True)
```

```python
df = pd.merge(
    sun_df.loc[:,["location","DateTime","太陽仰角", "日射量 (W/m²)","太陽方位角", "太陽能板日射量 (W/m²)"]],
    weather_df_10mins,
    left_on=["DateTime"],
    right_on=["DateTime"],
    how="left"
)
df
```

```python
df["日期"] = pd.to_datetime(df["DateTime"].dt.date)
df["時間"] = df["DateTime"].dt.hour.astype(str).str.zfill(2)
df["時段"] = df["DateTime"].dt.minute.astype(str).str.zfill(2)
```

```python
df
```

```python
df = pd.merge(
    df,
    weather_df_1hour,
    left_on=["日期","時間"],
    right_on=["日期","時間"],
    how="left",
    suffixes=["","_r"]
)
df
```

```python
df.to_csv("weather_sun_combine.csv",index=False)
```
