```python
eles = driver.find_elements(By.CLASS_NAME,"file-row.type-file")
for ele in eles:
    
    filename = ele.find_elements(By.CLASS_NAME,"is-block.name")[0].text
    ele.click()
    time.sleep(random.random())
```

```python
weather_file_path = r"Your_path"
filesnames = os.listdir(weather_file_path)
dflist = []
# filename = filesnames[0]
for filename in filesnames:
    zip_file_path = os.path.join(weather_file_path,filename)
    # 打開壓縮檔並讀取 CSV 文件
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # 獲取壓縮檔中的文件列表
        file_list = zip_ref.namelist()
        csv_file_name = file_list[0]
        # print("壓縮檔中的文件列表:", file_list)
        with zip_ref.open(csv_file_name) as csv_file:
        # 使用 pandas 讀取 CSV 文件
            df = pd.read_csv(csv_file)
            if "phenomenonTime" in list(df.columns):
                df.rename(columns={
                    "phenomenonTime":"DateTime"
                },
                inplace=True
                )
            #選取跟日照溫度有關的欄位
            columns_to_keep = [
            "StationId",
            "DateTime",
            "SunshineDuration",                # 日照時數
            "UVIndex",                         # 紫外線指數
            "AirTemperature",                  # 氣溫
            "DailyExtremeHighAirTemperature",  # 當日最高氣溫
            "DailyExtremeLowAirTemperature",   # 當日最低氣溫
            "RelativeHumidity",                # 相對濕度
            "VisibilityDescription",           # 能見度描述
            "StationLatitude",                 # 緯度
            "StationLongitude",                # 經度
            "StationAltitude"                  # 高度
        ]
        df = df.loc[df["StationId"]=="466990",columns_to_keep].reset_index(drop=True)  #選取花蓮氣象站
        print(filename)
        dflist.append(df)
```

```python
dfmerge = pd.concat(dflist,axis=0)
dfmerge
```

```python
dfmerge.drop_duplicates(inplace=True)
```

```python
dfmerge.drop_duplicates(subset=["DateTime"],keep="last",inplace=True)
```

```python
dfmerge.to_csv("10mins_weather_data.csv",index=False)
```
